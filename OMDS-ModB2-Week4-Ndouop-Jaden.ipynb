# 2. Weekly graph question

Below are a histogram and boxplot representation of the same data. A pharmacy is keeping a record of the prices of the drugs that it sells, and an administrator wants to know how much the more expensive drugs tend to cost, in the context of the other prices.

Please write a short explanation of the pros and cons of these two representations. Which would you choose? How would you modify the formatting, if at all, to make it more visually interesting, clear, or informative?

```python
import numpy as np
import pandas as pd

np.random.seed(0)
num_data = 100
data = np.exp(np.random.uniform(size = num_data) * 4)
df = pd.DataFrame(data.T, columns = ["data"])
```

```python
print("The 75th percentile is:", df.quantile(q = 0.75))
df.plot.hist()
```

```python
df.plot.box()
```

One pro of the histogram is that it shows how the drug prices are spread out. You can easily tell that most prices are low, and a few go really high, which makes the data look skewed. This makes it good for spotting patterns and seeing the overall shape of the data. A con is that it’s hard to tell exactly where the median or 75th percentile is just by looking at the bars.

One pro of the boxplot is that it clearly shows the median and quartiles, so you can quickly see what counts as an unusually high price. This is helpful for understanding how the more expensive drugs compare to the rest. A con is that it doesn’t show the full shape of the distribution like the histogram does, so you don’t get a sense of where the data is bunched up or spread out.

If I had to choose one for this situation, I’d pick the boxplot because it’s better for identifying which prices are high compared to most of the others. That seems more useful for the pharmacy administrator.

To make the boxplot more clear, I would add gridlines and labels to mark the median and percentiles, maybe even color the outliers differently so they stand out more. A title would help too.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("hypertension_dataset.csv") 

df['Hypertension'] = df['Hypertension'].str.strip().map({'Low': 0, 'High': 1})
df['Gender'] = df['Gender'].str.title()
df['Smoking_Status'] = df['Smoking_Status'].str.title()

sns.histplot(df['Age'], kde=True)
plt.title("Histogram + KDE of Age")
plt.show()

sns.violinplot(x='Hypertension', y='BMI', data=df)
plt.title("Violin Plot of BMI by Hypertension")
plt.show()

sns.boxplot(x='Hypertension', y='Glucose', data=df)
plt.title("Box Plot of Glucose by Hypertension")
plt.show()

sns.histplot(data=df, x='Cholesterol', hue='Hypertension', multiple='stack', bins=30)
plt.title("Grouped Histogram of Cholesterol by Hypertension")
plt.show()

sns.countplot(x='Smoking_Status', data=df)
plt.title("Bar Plot of Smoking Status")
plt.xticks(rotation=45)
plt.show()

print("Mean, Median, Mode:")
print(df[['Age', 'BMI', 'Glucose']].agg(['mean', 'median', lambda x: x.mode()[0]]))

print("\nMin and Max:")
print(df[['Age', 'BMI', 'Glucose']].agg(['min', 'max']))

Q1 = df['Glucose'].quantile(0.25)
Q3 = df['Glucose'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Glucose'] < Q1 - 1.5 * IQR) | (df['Glucose'] > Q3 + 1.5 * IQR)]
print(f"\nNumber of Glucose outliers: {len(outliers)}")

print("\nSkewness:")
print(df[['Age', 'BMI', 'Glucose']].skew())

sns.histplot(df['BMI'], bins=10)
plt.title("Histogram of BMI (10 bins)")
plt.show()

sns.histplot(df['BMI'], bins=30)
plt.title("Histogram of BMI (30 bins)")
plt.show()
```

One thing I noticed is that most of the distributions are skewed to the right, especially cholesterol and glucose. That means there are a few really high values pulling the data to the right. Age looks more normally distributed. This is helpful for seeing how the data is shaped and where most of the values are.

There are some outliers in the data too. Cholesterol and glucose have values that are far away from the others. These could be real or they could be errors, but either way, they stand out from the rest of the data.

I think this dataset is usable. The data seems clean, there’s a good amount of variation, and it makes sense based on what you’d expect from a health dataset. I didn’t see anything that would stop me from working with it.

If you change the number of bins in the histogram, it definitely changes how you see the data. When I used 10 bins, it looked pretty smooth, but with 30 bins I started to notice more small peaks and dips, especially in the glucose distribution. So bin size matters a lot when you're trying to spot patterns.

If I had to describe the overall distribution, I’d say some of the variables look normal, but most are skewed. That’s something to keep in mind for analysis or modeling later on.

# 4. Storytelling With Data graph

Reproduce any graph of your choice in p. 52-68 of the Storytelling With Data book as best you can.  (The second half of chapter two).  You do not have to get the exact data values right, just the overall look and feel.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Category': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E'],
    'Revenue': [23, 17, 35, 29, 12]
}

df = pd.DataFrame(data)
df = df.sort_values(by='Revenue', ascending=True)

plt.figure(figsize=(8, 5))
bars = plt.barh(df['Category'], df['Revenue'], color='#4C72B0')

for bar in bars:
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
             f"{bar.get_width()}", va='center', fontsize=10)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.tick_params(left=False, bottom=False)
plt.xticks([])

plt.title("Revenue by Product Category", fontsize=14, weight='bold')

plt.tight_layout()
plt.show()
```
